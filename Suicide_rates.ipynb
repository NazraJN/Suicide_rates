```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the datasets
female_suicide_rates = pd.read_csv('Age_adjusted_female_suicide_rates.csv')
male_suicide_rates = pd.read_csv('Age_adjusted_male_suicide_rates.csv')
```

```python
# check column features
print(female_suicide_rates.columns)
print(male_suicide_rates.columns)
```

```python
# summary of the dataset features
print(female_suicide_rates.info())
print(male_suicide_rates.info())
```

##**Female suicide rates**

```python
# preview first 20 columns
female_suicide_rates.head(20)
```

```python
# preview last 20 columns
female_suicide_rates.tail(20)
```

```python
# check for total missing values in each column
female_suicide_rates.isna().sum()
```

Observations:

- The *female suicide rates for selected countries* column has missing values

```python
# Drop the column with missing values
female_suicide_rates = female_suicide_rates.drop(columns=['Female suicide rates for selected countries, 1950-2005 (WHO (2015))'])
female_suicide_rates
```

```python
# Group data by year and calculate mean female suicide rates
mean_female_suicide_rates_by_year = female_suicide_rates.groupby('Year').mean()

# Plotting
plt.figure(figsize=(8, 6))
mean_female_suicide_rates_by_year['Female suicide rate (WHO (2005))'].plot(kind='line', color = 'blue')
plt.title('Female Suicide Rates by Year')
plt.xlabel('Year')
plt.ylabel('Suicide Rate')
plt.legend(['Female Suicide Rate'])
plt.tight_layout()
plt.show()
```

```python
# Sort the data by mean female suicide rates in descending order
mean_female_suicide_rates_by_year_sorted = mean_female_suicide_rates_by_year.sort_values(by='Female suicide rate (WHO (2005))', ascending=False)

# Get the top 10 years with the highest suicide rates
top_10_highest = mean_female_suicide_rates_by_year_sorted.head(10)

# Get the top 10 years with the lowest suicide rates
top_10_lowest = mean_female_suicide_rates_by_year_sorted.tail(10)

# Print the top 10 years with the highest suicide rates
print("Top 10 years with the highest female suicide rates:")
print(top_10_highest)

# Print the top 10 years with the lowest suicide rates
print("\nTop 10 years with the lowest female suicide rates:")
print(top_10_lowest)
```

```python
# Group data by country and calculate mean female suicide rates
mean_female_suicide_rates_by_country = female_suicide_rates.groupby(['Entity'])['Female suicide rate (WHO (2005))'].mean().reset_index()

# Sort the data by mean female suicide rates in descending order
mean_female_suicide_rates = mean_female_suicide_rates_by_country.sort_values(by='Female suicide rate (WHO (2005))', ascending=False)

# Select the top countries with the highest mean female suicide rates
highest_female_suicide_rates = mean_female_suicide_rates.head(10)

# Display the top countries
print(highest_female_suicide_rates)
```

```python
# Plotting
plt.figure(figsize=(12, 8))
plt.bar(highest_female_suicide_rates['Entity'], highest_female_suicide_rates['Female suicide rate (WHO (2005))'])
plt.title('Top 10 Countries with the Highest Female Suicide Rates (1950-2005)')
plt.xlabel('Country')
plt.ylabel('Female Suicide Rate')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Comparison of male and female suicide rates in countries with the highest female suicide rates

```python
# Merge highest female suicide rates with male suicide rates
merged_rates = pd.merge(highest_female_suicide_rates, male_suicide_rates, on='Entity', suffixes=('_female', '_male'))

# Plotting
plt.figure(figsize=(12, 8))
for index, row in merged_rates.iterrows():
    plt.bar([row['Entity'] + ' Female', row['Entity'] + ' Male'], [row['Female suicide rate (WHO (2005))'], row['Male suicide rate (WHO (2005))']], color=['green', 'blue'])

plt.title('Male and Female Suicide Rates in Countries with Highest Female Suicide Rates')
plt.ylabel('Suicide Rate')
#plt.legend(['Female Suicide Rate', 'Male Suicide Rate'])
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (12,10))
sns.boxplot(x='Year', y='Female suicide rate (WHO (2005))', data=female_suicide_rates)
plt.xlabel('Year')
plt.ylabel('Suicide Rate')
plt.title('Distribution of Suicide Rates by Year')
plt.xticks(rotation=90)
plt.show()
```

```python
# Select the top countries with the lowest mean female suicide rates
least_female_suicide_rates = mean_female_suicide_rates.tail(10)

# Display the top countries
print(least_female_suicide_rates)
```

```python
# Plotting
plt.figure(figsize=(15, 10))
plt.bar(least_female_suicide_rates['Entity'], least_female_suicide_rates['Female suicide rate (WHO (2005))'])
plt.title('Top 10 Countries with the Lowest Female Suicide Rates (1950-2005)')
plt.xlabel('Country')
plt.ylabel('Female Suicide Rate')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

**Male suicide rates**

```python
# preview first 20 columns
male_suicide_rates.head(20)
```

```python
# preview last 20 columns
male_suicide_rates.tail(20)
```

```python
# check for total missing values in each column
male_suicide_rates.isna().sum()
```

```python
# Drop the column with missing values
male_suicide_rates = male_suicide_rates.drop(columns=['Male suicide rates for selected countries, 1950-2005 (WHO (2005))'])
male_suicide_rates
```

```python
# Group data by year and calculate mean suicide rates
mean_male_suicide_rates_by_year = male_suicide_rates.groupby('Year').mean()

# Plotting
plt.figure(figsize=(12, 8))
mean_male_suicide_rates_by_year['Male suicide rate (WHO (2005))'].plot(kind='bar', color = 'orange')
plt.title('Male Suicide Rates by Year')
plt.xlabel('Year')
plt.ylabel('Suicide Rate')
plt.legend(['Male Suicide Rate'])
plt.tight_layout()
plt.show()
```

```python
# Sort the data by mean male suicide rates in descending order
mean_male_suicide_rates_by_year_sorted = mean_male_suicide_rates_by_year.sort_values(by='Male suicide rate (WHO (2005))', ascending=False)

# Get the top 10 years with the highest suicide rates
top_10_highest = mean_male_suicide_rates_by_year_sorted.head(10)

# Get the top 10 years with the lowest suicide rates
top_10_lowest = mean_male_suicide_rates_by_year_sorted.tail(10)

# Print the top 10 years with the highest suicide rates
print("Top 10 years with the highest male suicide rates:")
print(top_10_highest)

# Print the top 10 years with the lowest suicide rates
print("\nTop 10 years with the lowest male suicide rates:")
print(top_10_lowest)
```

```python
# Group data by country and calculate mean male suicide rates
mean_male_suicide_rates_by_country = male_suicide_rates.groupby(['Entity'])['Male suicide rate (WHO (2005))'].mean().reset_index()

# Sort the data by mean male suicide rate in descending order
sorted_male_suicide_rates = mean_male_suicide_rates_by_country.sort_values(by='Male suicide rate (WHO (2005))', ascending=False)

# Select the top countries with the highest mean male suicide rates
top_male_suicide_rates = sorted_male_suicide_rates.head(10)

# Display the top countries
print(top_male_suicide_rates)
```

```python
# Plotting
plt.figure(figsize=(15, 10))
plt.bar(top_male_suicide_rates['Entity'], top_male_suicide_rates['Male suicide rate (WHO (2005))'])
plt.title('Top 10 Countries with the Highest Male Suicide Rates (1950-2005)')
plt.xlabel('Country')
plt.ylabel('Male Suicide Rate')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Select the top countries with the lowest mean female suicide rates
least_male_suicide_rates = sorted_male_suicide_rates.tail(10)

# Display the top countries
print(least_male_suicide_rates)

# Plotting
plt.figure(figsize=(15, 10))
plt.bar(least_male_suicide_rates['Entity'], least_male_suicide_rates['Male suicide rate (WHO (2005))'])
plt.title('Top 10 Countries with the Lowest Male Suicide Rates (1950-2005)')
plt.xlabel('Country')
plt.ylabel('Male Suicide Rate')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Comparison of male and female suicide rates in countries with the lowest female suicide rates

```python
# Merge lowest female suicide rates with male suicide rates
merged_rates = pd.merge(least_female_suicide_rates, male_suicide_rates, on='Entity', suffixes=('_female', '_male'))

# Plotting
plt.figure(figsize=(12, 8))
for index, row in merged_rates.iterrows():
    plt.bar([row['Entity'] + ' Female', row['Entity'] + ' Male'], [row['Female suicide rate (WHO (2005))'], row['Male suicide rate (WHO (2005))']], color=['red', 'orange'])

plt.title('Male and Female Suicide Rates in Countries with Lowest Female Suicide Rates')
plt.ylabel('Suicide Rate')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Comparison of male and female suicide rates in countries with the highest male suicide rates

## Comparison of male and female suicide rates in countries with the lowest male suicide rates

```python
# Merge lowest male suicide rates with male suicide rates
merged_rates = pd.merge(least_male_suicide_rates, female_suicide_rates, on='Entity', suffixes=('_female', '_male'))

# Plotting
plt.figure(figsize=(12, 8))
for index, row in merged_rates.iterrows():
    plt.bar([row['Entity'] + ' Female', row['Entity'] + ' Male'], [row['Female suicide rate (WHO (2005))'], row['Male suicide rate (WHO (2005))']], color=['orange', 'green'])

plt.title('Male and Female Suicide Rates in Countries with Lowest Male Suicide Rates')
plt.ylabel('Suicide Rate')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
